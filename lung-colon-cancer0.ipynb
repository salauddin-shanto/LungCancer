```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import confusion_matrix, classification_report, roc_curve, auc
from sklearn.preprocessing import LabelBinarizer
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten, Dense, Dropout
from tensorflow.keras.callbacks import EarlyStopping
from sklearn.model_selection import StratifiedKFold
from keras.preprocessing import image
import os

# Directory paths
path = '/kaggle/input/lung-and-colon-cancer-histopathological-images/lung_colon_image_set/lung_image_sets'

# Set up classes
classes = ['lung_aca', 'lung_n', 'lung_scc']

# Prepare images and labels
def load_images(path, classes):
    images = []
    labels = []
    for label, class_name in enumerate(classes):
        class_path = os.path.join(path, class_name)
        for img_name in os.listdir(class_path):
            img_path = os.path.join(class_path, img_name)
            img = image.load_img(img_path, target_size=(128, 128))  # Resize all images
            img = image.img_to_array(img)
            images.append(img)
            labels.append(label)
    return np.array(images), np.array(labels)

# Load and preprocess data
X, y = load_images(path, classes)
X = X / 255.0  # Normalize the images

# Display random images from each class
def display_random_images(classes, path, num_images=3):
    fig, ax = plt.subplots(1, len(classes), figsize=(15, 5))
    fig.suptitle('Random Images for Each Class', fontsize=20)

    for i, class_name in enumerate(classes):
        class_path = os.path.join(path, class_name)
        images = os.listdir(class_path)
        random_images = np.random.choice(images, num_images, replace=False)

        # Display random images for the class
        for j, img_name in enumerate(random_images):
            img_path = os.path.join(class_path, img_name)
            img = image.load_img(img_path, target_size=(128, 128))
            img = image.img_to_array(img)
            ax[i].imshow(img)
            ax[i].axis('off')
        ax[i].set_title(class_name)

    plt.show()

# Display random images for each class
display_random_images(classes, path)

# Set up cross-validation
kfold = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)

# Initialize lists to store results
history_list = []
all_predictions = []
all_true_labels = []

# Model definition function
def create_model():
    model = Sequential([
        Conv2D(32, (3, 3), activation='relu', input_shape=(128, 128, 3)),
        MaxPooling2D(2, 2),
        Conv2D(64, (3, 3), activation='relu'),
        MaxPooling2D(2, 2),
        Flatten(),
        Dense(128, activation='relu'),
        Dropout(0.5),
        Dense(len(classes), activation='softmax')
    ])
    model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])
    return model

# Cross-validation training loop
for fold, (train_idx, val_idx) in enumerate(kfold.split(X, y)):
    print(f"Training fold {fold + 1}...")
    
    X_train, X_val = X[train_idx], X[val_idx]
    y_train, y_val = y[train_idx], y[val_idx]
    
    # One-hot encode labels before training the model
    lb = LabelBinarizer()
    y_train = lb.fit_transform(y_train)
    y_val = lb.transform(y_val)
    
    # Create model
    model = create_model()
    
    # Early stopping to avoid overfitting
    early_stopping = EarlyStopping(monitor='val_loss', patience=4, restore_best_weights=True)
    
    # Train the model
    history = model.fit(X_train, y_train, epochs=40, batch_size=32, validation_data=(X_val, y_val), 
                        callbacks=[early_stopping])
    
    history_list.append(history)

    # Collect predictions for later analysis
    y_pred = model.predict(X_val)
    y_pred_classes = np.argmax(y_pred, axis=1)
    all_predictions.append(y_pred_classes)
    all_true_labels.append(np.argmax(y_val, axis=1))

# Calculate overall accuracy and loss
average_accuracy = np.mean([history.history['accuracy'][-1] for history in history_list])
average_loss = np.mean([history.history['loss'][-1] for history in history_list])

# Print overall accuracy and loss
print(f"Average Accuracy: {average_accuracy:.4f}")
print(f"Average Loss: {average_loss:.4f}")

# Confusion Matrix
all_predictions = np.concatenate(all_predictions)
all_true_labels = np.concatenate(all_true_labels)
cm = confusion_matrix(all_true_labels, all_predictions)
plt.figure(figsize=(8, 6))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', xticklabels=classes, yticklabels=classes)
plt.title('Confusion Matrix')
plt.xlabel('Predicted')
plt.ylabel('True')
plt.savefig('/kaggle/working/confusion_matrix.png')  # Save the plot
plt.show()

# Classification report
print("Classification Report:\n", classification_report(all_true_labels, all_predictions, target_names=classes))

# ROC Curve (One vs Rest)
lb = LabelBinarizer()
y_bin = lb.fit_transform(y)  # Binarize the labels for ROC

fpr, tpr, _ = roc_curve(y_bin.ravel(), np.concatenate([model.predict(X[val_idx]).ravel() for _, val_idx in kfold.split(X, y)]).ravel())
roc_auc = auc(fpr, tpr)

plt.figure(figsize=(8, 6))
plt.plot(fpr, tpr, color='blue', label=f'AUC = {roc_auc:.2f}')
plt.plot([0, 1], [0, 1], color='gray', linestyle='--')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('ROC Curve')
plt.legend(loc='lower right')
plt.savefig('/kaggle/working/roc_curve.png')  # Save the plot
plt.show()

# Calculate the average training and validation accuracy and loss across folds
num_epochs = min([len(history.history['accuracy']) for history in history_list])

# Initialize arrays to accumulate results
avg_train_accuracy = np.zeros(num_epochs)
avg_val_accuracy = np.zeros(num_epochs)
avg_train_loss = np.zeros(num_epochs)
avg_val_loss = np.zeros(num_epochs)

for history in history_list:
    avg_train_accuracy += np.array(history.history['accuracy'][:num_epochs])
    avg_val_accuracy += np.array(history.history['val_accuracy'][:num_epochs])
    avg_train_loss += np.array(history.history['loss'][:num_epochs])
    avg_val_loss += np.array(history.history['val_loss'][:num_epochs])

# Calculate the mean across folds
avg_train_accuracy /= len(history_list)
avg_val_accuracy /= len(history_list)
avg_train_loss /= len(history_list)
avg_val_loss /= len(history_list)

# Plot the averaged training and validation accuracy
plt.figure(figsize=(10, 5))
plt.plot(avg_train_accuracy, label='Training Accuracy', color='blue')
plt.plot(avg_val_accuracy, label='Validation Accuracy', color='orange')
plt.title('Average Training and Validation Accuracy')
plt.xlabel('Epochs')
plt.ylabel('Accuracy')
plt.legend()
plt.savefig('/kaggle/working/avg_training_validation_accuracy.png')  # Save the plot
plt.show()

# Plot the averaged training and validation loss
plt.figure(figsize=(10, 5))
plt.plot(avg_train_loss, label='Training Loss', color='blue')
plt.plot(avg_val_loss, label='Validation Loss', color='orange')
plt.title('Average Training and Validation Loss')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.legend()
plt.savefig('/kaggle/working/avg_training_validation_loss.png')  # Save the plot
plt.show()


# Countplot for the distribution of classes
plt.figure(figsize=(8, 6))
sns.countplot(x=y, palette="Set2")
plt.title('Distribution of Classes')
plt.xlabel('Class')
plt.ylabel('Count')
plt.xticks(ticks=np.arange(len(classes)), labels=classes)
plt.savefig('/kaggle/working/countplot.png')  # Save the plot
plt.show()
```
